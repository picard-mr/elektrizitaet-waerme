# Isothermen

```python
%matplotlib notebook
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

p = []
V = np.linspace(10e-5, 0.01, 1e3) # Einheit: m^3
n = 1 # Einheit: mol
R = 8.314472 # Einheit: J/(K mol)




# Initialize figure
fig = plt.figure(figsize=(10, 6))
ax1 = fig.add_subplot(1, 2, 1)

# Multiple isotherms
T = np.linspace(100, 2000, 10) # Einheit: K
for i in np.arange(len(T)):
    p.append((n*R*T[i])/V) # Einheit: Pa
    plt.plot(V, p[i])
plt.xlim(1e-4,8e-4)
plt.show()




# Interactive isotherm
x = np.linspace(0, 2 * np.pi)
ax2 = fig.add_subplot(1, 2, 2)

T = 300 # Einheit: K
line, = ax2.plot(V, (n*R*T)/V)

def update(T = 300.0):
    line.set_ydata((n*R*T)/V)
    fig.canvas.draw_idle()

interact(update);


# TODO
"""
* Add comments
* Add axis descriptions
"""
```
