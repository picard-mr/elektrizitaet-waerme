## Isotherme

```python
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
%matplotlib notebook
```

```python
def pressure(V, T=300, n=1):
    return (n*8.314472*T)/V


n = 1 # Einheit: mol
R = 8.314472 # Einheit: J/(K mol)
T = 300 # Einheit: K

deltaV = 1e3
V = np.linspace(10e-5, 0.01, deltaV) # Einheit: m^3
p = pressure(V, T, n) # Einheit: Pa

# Integration limits
a_index = 50 # Will be used as V[50] = V[a_index]
b_index = 300

# Plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1)

isotherm = ax.plot(V, p)

# Make the shaded region
iV = np.linspace(V[a_index], V[b_index])
ip = pressure(iV, T, n)
verts = [(V[a_index], 0), *zip(iV, ip), (V[b_index], 0)]
poly = Polygon(verts, facecolor='0.9', edgecolor='0.5')
ax.add_patch(poly)

# Figure settings
ax.set_xlim(V[0], V[-1])
ax.set_ylim(p[-1], p[0])
plt.xlabel('Volume $V \,[m^3]$')
plt.ylabel('Pressure $p \,[Pa]$')

plt.show()
```

```python
# Integral as work for isotherms
W = np.trapz(p[a_index:b_index], x=V[a_index:b_index])
print("Numerical: " + "{:.2f}".format(W))

W_teo = n*R*T*np.log(V[a_index]/V[b_index])
print("Teorical: " + "{:.2f}".format(W_teo))
```

```python
# Make it interactive

p = []
V = np.linspace(10e-5, 0.01, 1e3) # Einheit: m^3
n = 1 # Einheit: mol
R = 8.314472 # Einheit: J/(K mol)
T = 300 # Einheit: K


# Interactive isotherm
x = np.linspace(0, 2 * np.pi)
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1)

T = 300 # Einheit: K
isotherm_1, = ax.plot(V, (n*R*T)/V)
isotherm_2 = isotherm_1

def update(T1 = 300.0, T2 = 400.0):
    isotherm_1.set_ydata((n*R*T1)/V)
    fig.canvas.draw_idle()
    isotherm_2.set_ydata((n*R*T2)/V)
    fig.canvas.draw_idle()

interact(update);
```

## Adiabatische Kurve

```python
def adiabatic_pressure(V, T=300, n=1):
    return (n*8.314472*T)/V


deltaV = 1e3
V = np.linspace(10e-5, 0.01, deltaV) # Einheit: m^3
p = adiabatic_pressure(V, T, n) # Einheit: Pa

# Integration limits
a_index = 50 # Will be used as V[50] = V[a_index]
b_index = 300

# Plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1)

isotherm = ax.plot(V, p)

# Make the shaded region
iV = np.linspace(V[a_index], V[b_index])
ip = pressure(iV, T, n)
verts = [(V[a_index], 0), *zip(iV, ip), (V[b_index], 0)]
poly = Polygon(verts, facecolor='0.9', edgecolor='0.5')
ax.add_patch(poly)

# Figure settings
ax.set_xlim(V[0], V[-1])
ax.set_ylim(p[-1], p[0])
plt.xlabel('Volume $V \,[m^3]$')
plt.ylabel('Pressure $p \,[Pa]$')

plt.show()
```
